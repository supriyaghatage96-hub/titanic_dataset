```python
# preprocessing part
# in this part we will apply standardscaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train[:5]
```

```python
x_train = pd.DataFrame(x_train, columns=x.columns)
x_test = pd.DataFrame(x_test, columns=x.columns)
x_train.head()
```

```python
import numpy as np
import pandas as pd



# logistic regression model
# used when oupput is binary

from sklearn.linear_model import LogisticRegression

# creting model

model= LogisticRegression()

# training model
# Fill NaN values in the 'embarked' column of x_train and x_test
# Since the mapping resulted in all NaNs, and 'S' (0) was likely the mode,
# filling with 0 is a reasonable approach.
x_train['embarked'].fillna(0, inplace=True)
x_test['embarked'].fillna(0, inplace=True)

model.fit(x_train,y_train)
# here the ml actually learns pattern
```

```python
# accuracy check

from sklearn.metrics import accuracy_score

y_pred_train = model.predict(x_train)
y_pred = model.predict(x_test)

train_acc = accuracy_score(y_train,y_pred_train)
test_acc = accuracy_score(y_test,y_pred)

print(train_acc,test_acc)

# accuracy alone is not enough for classifaction
# if 90 died, 10 survived but the model says everyone died
# then model wilp be useless

# here is why are using confusion matrix
```

```python
# proper fit - high train and high test,with a small gap
# overfit - big gap (train>>test)
# underfit - both train and test perform poorly
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
# [tn fp]
# [fn tp]
# tn = true negative - predicted died and actually died
# fp = false positive - predicted survived and actually died
# fn = false negative - predicted died and actually survived
# tp = true positive - predicted survived and actually survived
```

```python
# confusion matrix is to denote where the model got confused and
# where the model made dangerous mistake

# a confusion matrix is a table used only in classfication that
# compare actual and predicted classes, showing the
# number of correct and incorrect predictions.it provides much
```

```python
# visual confusion matrix

sns.heatmap(cm,annot=True,cmap='Blues')
plt.title('confusion matrix')
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
```
